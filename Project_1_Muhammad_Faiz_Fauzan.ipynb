```
from sympy import Float
# operasi bilangan yang memiliki koma (Float)

variabel1 = Float(2.3, 2) # fungsi angka 2 dibelang koma adalah untuk membatasi agar input nya hanya 2 digit
variabel2 = Float(3.2, 2)

penjumlahan = variabel1 + variabel2
print(penjumlahan)

# fungsi angka 2 dibelang koma adalah untuk membatasi agar input nya hanya 2 digit
print(Float(1.6666666,2))
print(Float(2.33333,2))

Float(20) # maka saat di print akan menghasilkan output 20.0 (bilangan integer menjadi float, 20 menjadi mempunyai angka di belakang koma)
Float(6.78,3) # fungsi angka 3 dibelang koma adalah untuk membatasi agar input nya hanya 3 digit
```

```
from sympy import Integer

# operasi bilangan bulat (Integer), dapat ditulis int
Integer(3.4) # saat di print akan menghasilkan output 3, karena bilangan dloat tersebut akan diubah menjadi integer dan dibulatkan kebawah menjadi 3
Integer(2/7) # saat di print akan menghasilkan output 0, karena hasil dari 2/7 adalah 0.28, dan di bulatkan menjadi 0


variabel1 = Integer(2)
variabel2 = Integer(8)
variabel3 = Integer(7)
variabel4 = Integer(2)

#penjumlahan
penjumlahan = variabel1 + variabel2
print(penjumlahan)

#pengurangan
pengurangan = variabel2 - variabel1
print(pengurangan)

#pengalian
kali = variabel1 * variabel2
print(kali)

#pembagian
pembagian = variabel2 / variabel1
print(pembagian)

#modulus
modulus = variabel3 % variabel4
print(modulus)
```
