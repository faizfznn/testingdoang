3. apakah fungsi dari return None

Dalam bahasa pemrograman Python, ketika sebuah fungsi tidak memiliki pernyataan return atau memiliki pernyataan return tanpa nilai yang dikembalikan (misalnya, return None), maka secara default, fungsi tersebut mengembalikan None.

Ini memberitahukan bahwa fungsi tersebut memang dimaksudkan untuk mengembalikan nilai untuk digunakan nanti, dan dalam hal ini fungsi tersebut mengembalikan None. Nilai ini Nonekemudian dapat digunakan di tempat lain. return Nonetidak pernah digunakan jika tidak ada kemungkinan nilai kembalian lain dari fungsi tersebut.
